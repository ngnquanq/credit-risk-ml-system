```python
train_df = pl.read_csv('../data/application_train.csv')
test_df = pl.read_csv('../data/application_test.csv')
```

```python
final_ext = pl.concat([train_ext, test_ext])
final_ext.write_csv('../data/application_ext.csv')
```
