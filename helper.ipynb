```python
import os
import pandas as pd
from sqlalchemy import create_engine

# DB connection config
DB_NAME = 'homecredit'
DB_USER = 'quang_user'
DB_PASSWORD = 'quang2011'
DB_HOST = 'localhost'
DB_PORT = 5432

# CSV folder
CSV_FOLDER = 'data/'

# Connect to PostgreSQL using SQLAlchemy engine
engine = create_engine(f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}')

# List all CSV files
csv_files = [f for f in os.listdir(CSV_FOLDER) if f.endswith('.csv')]

for csv_file in csv_files:
    table_name = os.path.splitext(csv_file)[0].lower()
    file_path = os.path.join(CSV_FOLDER, csv_file)

    print(f"📦 Importing: {file_path} → {table_name}")

    # Load CSV
    df = pd.read_csv(file_path)

    # Clean up column names
    df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

    with engine.begin() as connection:
        df.to_sql(
            name=table_name,
            con=connection,
            if_exists='replace',
            index=False
        )


    print(f"✅ {len(df)} rows inserted into table '{table_name}'")

print("🎉 All CSVs loaded into PostgreSQL.")
```
